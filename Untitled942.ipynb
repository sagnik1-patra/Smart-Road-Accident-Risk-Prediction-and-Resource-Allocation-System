```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


# ============================================================
# Paths
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Road Accident Risk Prediction and Resource Allocation System"
CSV_PATH = os.path.join(BASE_DIR, "Transport_2024_Annexure_3.csv")

os.makedirs(BASE_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(CSV_PATH)

df.columns = (
    df.columns
    .str.strip()
    .str.replace("\n", " ")
    .str.replace("  ", " ")
)

state_col = "State/UT"

target_col = "State / UT - wise Total Number of Persons Killed in Road Accidents during - 2024 - Number"

rank_col = "State / UT - wise Total Number of Persons Killed in Road Accidents during - 2024 - Rank"


# ============================================================
# Data Cleaning
# ============================================================

for col in df.columns:
    if col != state_col:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace("%", "", regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=[target_col])

numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())


# ============================================================
# Feature Engineering
# ============================================================

df["Fatality_Growth_2021_2024"] = (
    (df[target_col] -
     df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2021"])
    /
    (df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2021"] + 1)
) * 100

df["Fatality_Growth_2023_2024"] = (
    (df[target_col] -
     df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2023"])
    /
    (df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2023"] + 1)
) * 100

df["Road_Safety_Risk_Index"] = (
    df["Share of States / UTs in Total Number of Persons Killed in Road Accidents - 2024"] * 0.30
    + df["Total Number of Persons Killed in Road Accidents Per Lakh Population - 2024"] * 0.25
    + df["Total Number of Persons Killed in Road Accidents per 10,000 Vehicles - 2022"] * 0.20
    + df["Total Number of Persons Killed in Road Accidents per 10,000 Km of Roads - 2022 (P)"] * 0.25
)

df["Risk_Level"] = pd.qcut(
    df["Road_Safety_Risk_Index"],
    q=3,
    labels=["Low", "Medium", "High"]
)

df["Recommended_Ambulances"] = np.ceil(df[target_col] / 500).astype(int)
df["Recommended_Highway_Patrols"] = np.ceil(df["Road_Safety_Risk_Index"] / 100).astype(int)
df["Recommended_Trauma_Centers"] = np.ceil(df[target_col] / 3000).astype(int)


# ============================================================
# Features and Targets
# ============================================================

drop_cols = [
    state_col,
    target_col,
    rank_col,
    "Risk_Level"
]

X = df.drop(columns=drop_cols, errors="ignore")
y_reg = df[target_col]

label_encoder = LabelEncoder()
y_cls = label_encoder.fit_transform(df["Risk_Level"])


# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_reg,
    y_cls,
    test_size=0.25,
    random_state=42,
    stratify=y_cls
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# ============================================================
# Random Forest Regression
# ============================================================

rf_regressor = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf_regressor.fit(X_train_scaled, y_reg_train)

rf_reg_pred = rf_regressor.predict(X_test_scaled)

rf_reg_metrics = {
    "Model": "Random Forest Regressor",
    "MAE": mean_absolute_error(y_reg_test, rf_reg_pred),
    "RMSE": np.sqrt(mean_squared_error(y_reg_test, rf_reg_pred)),
    "R2_Score": r2_score(y_reg_test, rf_reg_pred)
}


# ============================================================
# Random Forest Classification
# ============================================================

rf_classifier = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)

rf_classifier.fit(X_train_scaled, y_cls_train)

rf_cls_pred = rf_classifier.predict(X_test_scaled)

rf_cls_accuracy = accuracy_score(y_cls_test, rf_cls_pred)

cls_report = classification_report(
    y_cls_test,
    rf_cls_pred,
    target_names=label_encoder.classes_,
    output_dict=True,
    zero_division=0
)

cm = confusion_matrix(y_cls_test, rf_cls_pred)


# ============================================================
# ANN Regression Model
# ============================================================

ann_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

ann_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=25,
    restore_best_weights=True
)

history = ann_model.fit(
    X_train_scaled,
    y_reg_train,
    validation_split=0.2,
    epochs=300,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ann_pred = ann_model.predict(X_test_scaled).flatten()

ann_metrics = {
    "Model": "ANN Regressor",
    "MAE": mean_absolute_error(y_reg_test, ann_pred),
    "RMSE": np.sqrt(mean_squared_error(y_reg_test, ann_pred)),
    "R2_Score": r2_score(y_reg_test, ann_pred)
}


# ============================================================
# Save Models
# ============================================================

ann_model.save(os.path.join(BASE_DIR, "road_accident_ann_model.h5"))

with open(os.path.join(BASE_DIR, "road_accident_ml_models.pkl"), "wb") as f:
    pickle.dump({
        "scaler": scaler,
        "label_encoder": label_encoder,
        "rf_regressor": rf_regressor,
        "rf_classifier": rf_classifier,
        "features": list(X.columns)
    }, f)


# ============================================================
# Save YAML and JSON
# ============================================================

config = {
    "project_name": "Smart Road Accident Risk Prediction and Resource Allocation System",
    "dataset_path": CSV_PATH,
    "target_column": target_col,
    "classification_target": "Risk_Level",
    "features": list(X.columns),
    "risk_classes": list(label_encoder.classes_)
}

with open(os.path.join(BASE_DIR, "road_accident_config.yaml"), "w") as f:
    yaml.dump(config, f, sort_keys=False)

metrics_json = {
    "random_forest_regression": {
        "MAE": float(rf_reg_metrics["MAE"]),
        "RMSE": float(rf_reg_metrics["RMSE"]),
        "R2_Score": float(rf_reg_metrics["R2_Score"])
    },
    "ann_regression": {
        "MAE": float(ann_metrics["MAE"]),
        "RMSE": float(ann_metrics["RMSE"]),
        "R2_Score": float(ann_metrics["R2_Score"])
    },
    "random_forest_classification": {
        "Accuracy": float(rf_cls_accuracy),
        "Classification_Report": cls_report,
        "Confusion_Matrix": cm.tolist()
    }
}

with open(os.path.join(BASE_DIR, "road_accident_metrics.json"), "w") as f:
    json.dump(metrics_json, f, indent=4)


# ============================================================
# Result CSV
# ============================================================

result_df = pd.DataFrame({
    "Actual_2024_Fatalities": y_reg_test.values,
    "RF_Predicted_Fatalities": rf_reg_pred.round(0).astype(int),
    "ANN_Predicted_Fatalities": ann_pred.round(0).astype(int),
    "Actual_Risk_Level": label_encoder.inverse_transform(y_cls_test),
    "Predicted_Risk_Level": label_encoder.inverse_transform(rf_cls_pred)
})

result_csv_path = os.path.join(BASE_DIR, "road_accident_result.csv")
result_df.to_csv(result_csv_path, index=False)


# ============================================================
# Prediction CSV
# ============================================================

full_scaled = scaler.transform(X)

df["Predicted_2025_Fatalities_RF"] = rf_regressor.predict(full_scaled).round(0).astype(int)

df["Predicted_2025_Fatalities_ANN"] = ann_model.predict(full_scaled).flatten().round(0).astype(int)

df["Predicted_Risk_Level"] = label_encoder.inverse_transform(
    rf_classifier.predict(full_scaled)
)

prediction_csv_path = os.path.join(BASE_DIR, "road_accident_prediction.csv")
df.to_csv(prediction_csv_path, index=False)


# ============================================================
# Accuracy Graph
# ============================================================

accuracy_graph_path = os.path.join(BASE_DIR, "road_accident_accuracy_graph.png")

plt.figure(figsize=(7, 5))
plt.bar(["RF Classifier Accuracy"], [rf_cls_accuracy * 100])
plt.ylabel("Accuracy (%)")
plt.title("Road Accident Risk Classification Accuracy")
plt.ylim(0, 100)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()


# ============================================================
# Confusion Matrix Heatmap
# ============================================================

heatmap_path = os.path.join(BASE_DIR, "road_accident_confusion_matrix_heatmap.png")

plt.figure(figsize=(7, 5))
plt.imshow(cm)
plt.title("Confusion Matrix Heatmap")
plt.colorbar()

classes = label_encoder.classes_
plt.xticks(np.arange(len(classes)), classes)
plt.yticks(np.arange(len(classes)), classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()


# ============================================================
# Model Comparison Graph
# ============================================================

comparison_graph_path = os.path.join(BASE_DIR, "road_accident_comparison_graph.png")

comparison_df = pd.DataFrame([
    rf_reg_metrics,
    ann_metrics
])

comparison_csv_path = os.path.join(BASE_DIR, "road_accident_comparison.csv")
comparison_df.to_csv(comparison_csv_path, index=False)

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["R2_Score"])
plt.ylabel("R2 Score")
plt.title("Model Comparison Using R2 Score")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()


# ============================================================
# Result Graph: Actual vs Predicted
# ============================================================

result_graph_path = os.path.join(BASE_DIR, "road_accident_result_graph.png")

plt.figure(figsize=(9, 5))
plt.plot(result_df["Actual_2024_Fatalities"].values, marker="o", label="Actual 2024")
plt.plot(result_df["RF_Predicted_Fatalities"].values, marker="o", label="RF Predicted")
plt.plot(result_df["ANN_Predicted_Fatalities"].values, marker="o", label="ANN Predicted")
plt.xlabel("Test Samples")
plt.ylabel("Fatalities")
plt.title("Actual vs Predicted Road Accident Fatalities")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()


# ============================================================
# Prediction Graph: State-wise Predicted Fatalities
# ============================================================

prediction_graph_path = os.path.join(BASE_DIR, "road_accident_prediction_graph.png")

plot_df = df.sort_values("Predicted_2025_Fatalities_RF", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(plot_df[state_col], plot_df["Predicted_2025_Fatalities_RF"])
plt.xlabel("State/UT")
plt.ylabel("Predicted 2025 Fatalities")
plt.title("Top 15 States/UTs by Predicted 2025 Road Accident Fatalities")
plt.xticks(rotation=75)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()


# ============================================================
# Resource Allocation Graph
# ============================================================

resource_graph_path = os.path.join(BASE_DIR, "road_accident_resource_allocation_graph.png")

resource_df = df.sort_values("Recommended_Ambulances", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(resource_df[state_col], resource_df["Recommended_Ambulances"])
plt.xlabel("State/UT")
plt.ylabel("Recommended Ambulances")
plt.title("Top 15 States/UTs by Recommended Ambulance Allocation")
plt.xticks(rotation=75)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(resource_graph_path, dpi=300)
plt.show()


# ============================================================
# Risk Distribution Graph
# ============================================================

risk_graph_path = os.path.join(BASE_DIR, "road_accident_risk_distribution_graph.png")

risk_counts = df["Predicted_Risk_Level"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(risk_counts.index, risk_counts.values)
plt.xlabel("Risk Level")
plt.ylabel("Number of States/UTs")
plt.title("Predicted Road Accident Risk Distribution")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(risk_graph_path, dpi=300)
plt.show()


# ============================================================
# Final Print
# ============================================================

print("\nAll files saved successfully in:")
print(BASE_DIR)

print("\nSaved Files:")
print("1. road_accident_ann_model.h5")
print("2. road_accident_ml_models.pkl")
print("3. road_accident_config.yaml")
print("4. road_accident_metrics.json")
print("5. road_accident_result.csv")
print("6. road_accident_prediction.csv")
print("7. road_accident_comparison.csv")
print("8. road_accident_accuracy_graph.png")
print("9. road_accident_confusion_matrix_heatmap.png")
print("10. road_accident_comparison_graph.png")
print("11. road_accident_result_graph.png")
print("12. road_accident_prediction_graph.png")
print("13. road_accident_resource_allocation_graph.png")
print("14. road_accident_risk_distribution_graph.png")

print("\nModel Metrics:")
print("RF Regression:", rf_reg_metrics)
print("ANN Regression:", ann_metrics)
print("RF Classification Accuracy:", rf_cls_accuracy)

print("\nSample Prediction Output:")
print(
    df[
        [
            state_col,
            target_col,
            "Road_Safety_Risk_Index",
            "Risk_Level",
            "Predicted_2025_Fatalities_RF",
            "Predicted_2025_Fatalities_ANN",
            "Predicted_Risk_Level",
            "Recommended_Ambulances",
            "Recommended_Highway_Patrols",
            "Recommended_Trauma_Centers"
        ]
    ].head()
)
```
