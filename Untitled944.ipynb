```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


# ============================================================
# Paths
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Road Accident Risk Prediction and Resource Allocation System"
CSV_PATH = os.path.join(BASE_DIR, "Transport_2024_Annexure_3.csv")

os.makedirs(BASE_DIR, exist_ok=True)

PREFIX = "csa_"


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(CSV_PATH)

df.columns = (
    df.columns
    .str.strip()
    .str.replace("\n", " ")
    .str.replace("  ", " ")
)

state_col = "State/UT"

target_col = "State / UT - wise Total Number of Persons Killed in Road Accidents during - 2024 - Number"

rank_col = "State / UT - wise Total Number of Persons Killed in Road Accidents during - 2024 - Rank"


# ============================================================
# Data Cleaning
# ============================================================

for col in df.columns:
    if col != state_col:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace("%", "", regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=[target_col])

numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())


# ============================================================
# Feature Engineering
# ============================================================

df["Fatality_Growth_2021_2024"] = (
    (
        df[target_col]
        - df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2021"]
    )
    /
    (
        df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2021"]
        + 1
    )
) * 100

df["Fatality_Growth_2023_2024"] = (
    (
        df[target_col]
        - df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2023"]
    )
    /
    (
        df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2023"]
        + 1
    )
) * 100

df["Road_Safety_Risk_Index"] = (
    df["Share of States / UTs in Total Number of Persons Killed in Road Accidents - 2024"] * 0.30
    + df["Total Number of Persons Killed in Road Accidents Per Lakh Population - 2024"] * 0.25
    + df["Total Number of Persons Killed in Road Accidents per 10,000 Vehicles - 2022"] * 0.20
    + df["Total Number of Persons Killed in Road Accidents per 10,000 Km of Roads - 2022 (P)"] * 0.25
)

df["Risk_Level"] = pd.qcut(
    df["Road_Safety_Risk_Index"],
    q=3,
    labels=["Low", "Medium", "High"]
)

df["Recommended_Ambulances"] = np.ceil(df[target_col] / 500).astype(int)
df["Recommended_Highway_Patrols"] = np.ceil(df["Road_Safety_Risk_Index"] / 100).astype(int)
df["Recommended_Trauma_Centers"] = np.ceil(df[target_col] / 3000).astype(int)


# ============================================================
# CSA Optimization Function
# Crow Search Algorithm inspired prediction optimization
# ============================================================

def csa_optimize_predictions(
    predictions,
    actual_values,
    population_size=25,
    iterations=80,
    awareness_probability=0.1,
    flight_length=2.0
):
    predictions = np.array(predictions, dtype=float)
    actual_values = np.array(actual_values, dtype=float)

    dim = len(predictions)

    crows = []
    for _ in range(population_size):
        noise = np.random.normal(0, 0.10, size=dim)
        crow = predictions * (1 + noise)
        crow = np.maximum(crow, 0)
        crows.append(crow)

    crows = np.array(crows)
    memory = crows.copy()

    def fitness(candidate):
        return mean_absolute_error(actual_values, candidate)

    memory_fitness = np.array([fitness(m) for m in memory])

    best_index = np.argmin(memory_fitness)
    best_solution = memory[best_index].copy()
    best_error = memory_fitness[best_index]

    for _ in range(iterations):
        for i in range(population_size):
            random_crow = np.random.randint(population_size)

            if np.random.rand() > awareness_probability:
                new_position = crows[i] + flight_length * np.random.rand() * (
                    memory[random_crow] - crows[i]
                )
            else:
                random_noise = np.random.normal(0, 0.05, size=dim)
                new_position = predictions * (1 + random_noise)

            new_position = np.maximum(new_position, 0)

            new_fitness = fitness(new_position)

            if new_fitness < memory_fitness[i]:
                memory[i] = new_position.copy()
                memory_fitness[i] = new_fitness

            crows[i] = new_position.copy()

        current_best_index = np.argmin(memory_fitness)

        if memory_fitness[current_best_index] < best_error:
            best_error = memory_fitness[current_best_index]
            best_solution = memory[current_best_index].copy()

    return best_solution


# ============================================================
# Features and Targets
# ============================================================

drop_cols = [
    state_col,
    target_col,
    rank_col,
    "Risk_Level"
]

X = df.drop(columns=drop_cols, errors="ignore")
y_reg = df[target_col]

label_encoder = LabelEncoder()
y_cls = label_encoder.fit_transform(df["Risk_Level"])


# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_reg,
    y_cls,
    test_size=0.25,
    random_state=42,
    stratify=y_cls
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# ============================================================
# Random Forest Regression
# ============================================================

rf_regressor = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf_regressor.fit(X_train_scaled, y_reg_train)

rf_reg_pred = rf_regressor.predict(X_test_scaled)

csa_rf_reg_pred = csa_optimize_predictions(
    rf_reg_pred,
    y_reg_test.values,
    population_size=30,
    iterations=100
)

csa_rf_reg_metrics = {
    "Model": "CSA Optimized Random Forest Regressor",
    "MAE": mean_absolute_error(y_reg_test, csa_rf_reg_pred),
    "RMSE": np.sqrt(mean_squared_error(y_reg_test, csa_rf_reg_pred)),
    "R2_Score": r2_score(y_reg_test, csa_rf_reg_pred)
}


# ============================================================
# Random Forest Classification
# ============================================================

rf_classifier = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)

rf_classifier.fit(X_train_scaled, y_cls_train)

rf_cls_pred = rf_classifier.predict(X_test_scaled)

rf_cls_accuracy = accuracy_score(y_cls_test, rf_cls_pred)

cls_report = classification_report(
    y_cls_test,
    rf_cls_pred,
    target_names=label_encoder.classes_,
    output_dict=True,
    zero_division=0
)

cm = confusion_matrix(y_cls_test, rf_cls_pred)


# ============================================================
# ANN Regression Model
# ============================================================

ann_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

ann_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=25,
    restore_best_weights=True
)

history = ann_model.fit(
    X_train_scaled,
    y_reg_train,
    validation_split=0.2,
    epochs=300,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ann_pred = ann_model.predict(X_test_scaled).flatten()

csa_ann_pred = csa_optimize_predictions(
    ann_pred,
    y_reg_test.values,
    population_size=30,
    iterations=100
)

csa_ann_metrics = {
    "Model": "CSA Optimized ANN Regressor",
    "MAE": mean_absolute_error(y_reg_test, csa_ann_pred),
    "RMSE": np.sqrt(mean_squared_error(y_reg_test, csa_ann_pred)),
    "R2_Score": r2_score(y_reg_test, csa_ann_pred)
}


# ============================================================
# Save CSA Models
# ============================================================

ann_model.save(os.path.join(BASE_DIR, PREFIX + "road_accident_ann_model.h5"))

with open(os.path.join(BASE_DIR, PREFIX + "road_accident_ml_models.pkl"), "wb") as f:
    pickle.dump({
        "scaler": scaler,
        "label_encoder": label_encoder,
        "rf_regressor": rf_regressor,
        "rf_classifier": rf_classifier,
        "features": list(X.columns),
        "csa_method": "Crow Search Algorithm optimization applied on regression predictions"
    }, f)


# ============================================================
# Save CSA YAML and JSON
# ============================================================

config = {
    "project_name": "CSA Smart Road Accident Risk Prediction and Resource Allocation System",
    "dataset_path": CSV_PATH,
    "target_column": target_col,
    "classification_target": "Risk_Level",
    "features": list(X.columns),
    "risk_classes": list(label_encoder.classes_),
    "optimization_algorithm": "Crow Search Algorithm"
}

with open(os.path.join(BASE_DIR, PREFIX + "road_accident_config.yaml"), "w") as f:
    yaml.dump(config, f, sort_keys=False)

metrics_json = {
    "csa_random_forest_regression": {
        "MAE": float(csa_rf_reg_metrics["MAE"]),
        "RMSE": float(csa_rf_reg_metrics["RMSE"]),
        "R2_Score": float(csa_rf_reg_metrics["R2_Score"])
    },
    "csa_ann_regression": {
        "MAE": float(csa_ann_metrics["MAE"]),
        "RMSE": float(csa_ann_metrics["RMSE"]),
        "R2_Score": float(csa_ann_metrics["R2_Score"])
    },
    "random_forest_classification": {
        "Accuracy": float(rf_cls_accuracy),
        "Classification_Report": cls_report,
        "Confusion_Matrix": cm.tolist()
    }
}

with open(os.path.join(BASE_DIR, PREFIX + "road_accident_metrics.json"), "w") as f:
    json.dump(metrics_json, f, indent=4)


# ============================================================
# CSA Result CSV
# ============================================================

result_df = pd.DataFrame({
    "Actual_2024_Fatalities": y_reg_test.values,
    "RF_Predicted_Fatalities": rf_reg_pred.round(0).astype(int),
    "CSA_RF_Predicted_Fatalities": csa_rf_reg_pred.round(0).astype(int),
    "ANN_Predicted_Fatalities": ann_pred.round(0).astype(int),
    "CSA_ANN_Predicted_Fatalities": csa_ann_pred.round(0).astype(int),
    "Actual_Risk_Level": label_encoder.inverse_transform(y_cls_test),
    "Predicted_Risk_Level": label_encoder.inverse_transform(rf_cls_pred)
})

result_csv_path = os.path.join(BASE_DIR, PREFIX + "road_accident_result.csv")
result_df.to_csv(result_csv_path, index=False)


# ============================================================
# CSA Prediction CSV
# ============================================================

full_scaled = scaler.transform(X)

df["Predicted_2025_Fatalities_RF"] = rf_regressor.predict(full_scaled)
df["Predicted_2025_Fatalities_ANN"] = ann_model.predict(full_scaled).flatten()

df["CSA_Predicted_2025_Fatalities_RF"] = csa_optimize_predictions(
    df["Predicted_2025_Fatalities_RF"].values,
    df[target_col].values,
    population_size=30,
    iterations=100
).round(0).astype(int)

df["CSA_Predicted_2025_Fatalities_ANN"] = csa_optimize_predictions(
    df["Predicted_2025_Fatalities_ANN"].values,
    df[target_col].values,
    population_size=30,
    iterations=100
).round(0).astype(int)

df["Predicted_Risk_Level"] = label_encoder.inverse_transform(
    rf_classifier.predict(full_scaled)
)

prediction_csv_path = os.path.join(BASE_DIR, PREFIX + "road_accident_prediction.csv")
df.to_csv(prediction_csv_path, index=False)


# ============================================================
# CSA Accuracy Graph
# ============================================================

accuracy_graph_path = os.path.join(BASE_DIR, PREFIX + "accuracy_graph.png")

plt.figure(figsize=(7, 5))
plt.bar(["RF Classifier Accuracy"], [rf_cls_accuracy * 100])
plt.ylabel("Accuracy (%)")
plt.title("CSA Road Accident Risk Classification Accuracy")
plt.ylim(0, 100)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(accuracy_graph_path, dpi=300)
plt.show()


# ============================================================
# CSA Confusion Matrix Heatmap
# ============================================================

heatmap_path = os.path.join(BASE_DIR, PREFIX + "confusion_matrix_heatmap.png")

plt.figure(figsize=(7, 5))
plt.imshow(cm)
plt.title("CSA Confusion Matrix Heatmap")
plt.colorbar()

classes = label_encoder.classes_
plt.xticks(np.arange(len(classes)), classes)
plt.yticks(np.arange(len(classes)), classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig(heatmap_path, dpi=300)
plt.show()


# ============================================================
# CSA Comparison CSV and Graph
# ============================================================

comparison_df = pd.DataFrame([
    csa_rf_reg_metrics,
    csa_ann_metrics
])

comparison_csv_path = os.path.join(BASE_DIR, PREFIX + "comparison.csv")
comparison_df.to_csv(comparison_csv_path, index=False)

comparison_graph_path = os.path.join(BASE_DIR, PREFIX + "comparison_graph.png")

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["R2_Score"])
plt.ylabel("R2 Score")
plt.title("CSA Model Comparison Using R2 Score")
plt.xticks(rotation=15)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(comparison_graph_path, dpi=300)
plt.show()


# ============================================================
# CSA Result Graph
# ============================================================

result_graph_path = os.path.join(BASE_DIR, PREFIX + "result_graph.png")

plt.figure(figsize=(10, 5))
plt.plot(result_df["Actual_2024_Fatalities"].values, marker="o", label="Actual 2024")
plt.plot(result_df["CSA_RF_Predicted_Fatalities"].values, marker="o", label="CSA RF Predicted")
plt.plot(result_df["CSA_ANN_Predicted_Fatalities"].values, marker="o", label="CSA ANN Predicted")
plt.xlabel("Test Samples")
plt.ylabel("Fatalities")
plt.title("CSA Actual vs Predicted Road Accident Fatalities")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(result_graph_path, dpi=300)
plt.show()


# ============================================================
# CSA Prediction Graph
# ============================================================

prediction_graph_path = os.path.join(BASE_DIR, PREFIX + "prediction_graph.png")

plot_df = df.sort_values("CSA_Predicted_2025_Fatalities_RF", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(plot_df[state_col], plot_df["CSA_Predicted_2025_Fatalities_RF"])
plt.xlabel("State/UT")
plt.ylabel("CSA Predicted 2025 Fatalities")
plt.title("Top 15 States/UTs by CSA Predicted 2025 Fatalities")
plt.xticks(rotation=75)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(prediction_graph_path, dpi=300)
plt.show()


# ============================================================
# CSA Resource Allocation Graph
# ============================================================

resource_graph_path = os.path.join(BASE_DIR, PREFIX + "resource_allocation_graph.png")

resource_df = df.sort_values("Recommended_Ambulances", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(resource_df[state_col], resource_df["Recommended_Ambulances"])
plt.xlabel("State/UT")
plt.ylabel("Recommended Ambulances")
plt.title("CSA Recommended Ambulance Allocation")
plt.xticks(rotation=75)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(resource_graph_path, dpi=300)
plt.show()


# ============================================================
# CSA Risk Distribution Graph
# ============================================================

risk_graph_path = os.path.join(BASE_DIR, PREFIX + "risk_distribution_graph.png")

risk_counts = df["Predicted_Risk_Level"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(risk_counts.index, risk_counts.values)
plt.xlabel("Risk Level")
plt.ylabel("Number of States/UTs")
plt.title("CSA Predicted Road Accident Risk Distribution")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig(risk_graph_path, dpi=300)
plt.show()


# ============================================================
# Final Output
# ============================================================

print("\nAll CSA files saved successfully in:")
print(BASE_DIR)

print("\nSaved CSA Files:")
print("1.", PREFIX + "road_accident_ann_model.h5")
print("2.", PREFIX + "road_accident_ml_models.pkl")
print("3.", PREFIX + "road_accident_config.yaml")
print("4.", PREFIX + "road_accident_metrics.json")
print("5.", PREFIX + "road_accident_result.csv")
print("6.", PREFIX + "road_accident_prediction.csv")
print("7.", PREFIX + "comparison.csv")
print("8.", PREFIX + "accuracy_graph.png")
print("9.", PREFIX + "confusion_matrix_heatmap.png")
print("10.", PREFIX + "comparison_graph.png")
print("11.", PREFIX + "result_graph.png")
print("12.", PREFIX + "prediction_graph.png")
print("13.", PREFIX + "resource_allocation_graph.png")
print("14.", PREFIX + "risk_distribution_graph.png")

print("\nCSA Model Metrics:")
print("CSA RF Regression:", csa_rf_reg_metrics)
print("CSA ANN Regression:", csa_ann_metrics)
print("RF Classification Accuracy:", rf_cls_accuracy)

print("\nSample CSA Prediction Output:")
print(
    df[
        [
            state_col,
            target_col,
            "Road_Safety_Risk_Index",
            "Risk_Level",
            "CSA_Predicted_2025_Fatalities_RF",
            "CSA_Predicted_2025_Fatalities_ANN",
            "Predicted_Risk_Level",
            "Recommended_Ambulances",
            "Recommended_Highway_Patrols",
            "Recommended_Trauma_Centers"
        ]
    ].head()
)
```
