```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping


# ============================================================
# Paths
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Road Accident Risk Prediction and Resource Allocation System"
CSV_PATH = os.path.join(BASE_DIR, "Transport_2024_Annexure_3.csv")

os.makedirs(BASE_DIR, exist_ok=True)

print("Loading dataset from:", CSV_PATH)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(CSV_PATH)

print("\nDataset Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())


# ============================================================
# Clean Column Names
# ============================================================

df.columns = (
    df.columns
    .str.strip()
    .str.replace("\n", " ")
    .str.replace("  ", " ")
)

state_col = "State/UT"

target_col = "State / UT - wise Total Number of Persons Killed in Road Accidents during - 2024 - Number"

rank_col = "State / UT - wise Total Number of Persons Killed in Road Accidents during - 2024 - Rank"


# ============================================================
# Numeric Conversion
# ============================================================

for col in df.columns:
    if col != state_col:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.replace("%", "", regex=False)
            .str.strip()
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=[target_col])

numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())


# ============================================================
# Feature Engineering
# ============================================================

df["Fatality_Growth_2021_2024"] = (
    (df[target_col] -
     df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2021"])
    /
    (df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2021"] + 1)
) * 100

df["Fatality_Growth_2023_2024"] = (
    (df[target_col] -
     df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2023"])
    /
    (df["State / UT - wise Total Number of Persons Killed in Road Accidents during - 2023"] + 1)
) * 100

df["Road_Safety_Risk_Index"] = (
    df["Share of States / UTs in Total Number of Persons Killed in Road Accidents - 2024"] * 0.30
    + df["Total Number of Persons Killed in Road Accidents Per Lakh Population - 2024"] * 0.25
    + df["Total Number of Persons Killed in Road Accidents per 10,000 Vehicles - 2022"] * 0.20
    + df["Total Number of Persons Killed in Road Accidents per 10,000 Km of Roads - 2022 (P)"] * 0.25
)

df["Risk_Level"] = pd.qcut(
    df["Road_Safety_Risk_Index"],
    q=3,
    labels=["Low", "Medium", "High"]
)

df["Recommended_Ambulances"] = np.ceil(df[target_col] / 500).astype(int)
df["Recommended_Highway_Patrols"] = np.ceil(df["Road_Safety_Risk_Index"] / 100).astype(int)
df["Recommended_Trauma_Centers"] = np.ceil(df[target_col] / 3000).astype(int)

print("\nRisk Level Distribution:")
print(df["Risk_Level"].value_counts())


# ============================================================
# Prepare Features and Targets
# ============================================================

drop_cols = [
    state_col,
    target_col,
    rank_col,
    "Risk_Level"
]

X = df.drop(columns=drop_cols, errors="ignore")
y_reg = df[target_col]

label_encoder = LabelEncoder()
y_cls = label_encoder.fit_transform(df["Risk_Level"])


# ============================================================
# Correct Train-Test Split
# ============================================================

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_reg,
    y_cls,
    test_size=0.25,
    random_state=42,
    stratify=y_cls
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# ============================================================
# Random Forest Regression
# ============================================================

rf_regressor = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf_regressor.fit(X_train_scaled, y_reg_train)

reg_pred = rf_regressor.predict(X_test_scaled)

regression_metrics = {
    "MAE": float(mean_absolute_error(y_reg_test, reg_pred)),
    "RMSE": float(np.sqrt(mean_squared_error(y_reg_test, reg_pred))),
    "R2_Score": float(r2_score(y_reg_test, reg_pred))
}

print("\nRandom Forest Regression Metrics:")
print(regression_metrics)


# ============================================================
# Random Forest Classification
# ============================================================

rf_classifier = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced"
)

rf_classifier.fit(X_train_scaled, y_cls_train)

cls_pred = rf_classifier.predict(X_test_scaled)

classification_metrics = {
    "Accuracy": float(accuracy_score(y_cls_test, cls_pred)),
    "Classification_Report": classification_report(
        y_cls_test,
        cls_pred,
        target_names=label_encoder.classes_,
        output_dict=True,
        zero_division=0
    ),
    "Confusion_Matrix": confusion_matrix(y_cls_test, cls_pred).tolist()
}

print("\nRandom Forest Classification Accuracy:")
print(classification_metrics["Accuracy"])


# ============================================================
# ANN Regression Model
# ============================================================

ann_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

ann_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = ann_model.fit(
    X_train_scaled,
    y_reg_train,
    validation_split=0.2,
    epochs=300,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

ann_pred = ann_model.predict(X_test_scaled).flatten()

ann_metrics = {
    "MAE": float(mean_absolute_error(y_reg_test, ann_pred)),
    "RMSE": float(np.sqrt(mean_squared_error(y_reg_test, ann_pred))),
    "R2_Score": float(r2_score(y_reg_test, ann_pred))
}

print("\nANN Regression Metrics:")
print(ann_metrics)


# ============================================================
# Save H5, PKL, YAML, JSON
# ============================================================

h5_path = os.path.join(BASE_DIR, "road_accident_ann_model.h5")
pkl_path = os.path.join(BASE_DIR, "road_accident_ml_models.pkl")
yaml_path = os.path.join(BASE_DIR, "road_accident_config.yaml")
json_path = os.path.join(BASE_DIR, "road_accident_metrics.json")

ann_model.save(h5_path)

with open(pkl_path, "wb") as f:
    pickle.dump({
        "scaler": scaler,
        "label_encoder": label_encoder,
        "rf_regressor": rf_regressor,
        "rf_classifier": rf_classifier,
        "features": list(X.columns)
    }, f)

config = {
    "project_name": "Smart Road Accident Risk Prediction and Resource Allocation System",
    "dataset_path": CSV_PATH,
    "target_column": target_col,
    "classification_target": "Risk_Level",
    "features": list(X.columns),
    "risk_classes": list(label_encoder.classes_),
    "models": {
        "random_forest_regressor": "Predicts future accident fatalities",
        "random_forest_classifier": "Classifies state risk level",
        "ann_model": "Deep learning regression model saved as H5"
    },
    "outputs": {
        "h5_model": h5_path,
        "pkl_models": pkl_path,
        "yaml_config": yaml_path,
        "json_metrics": json_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

metrics = {
    "random_forest_regression": regression_metrics,
    "random_forest_classification": classification_metrics,
    "ann_regression": ann_metrics
}

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)


# ============================================================
# Full Dataset Prediction
# ============================================================

full_scaled = scaler.transform(X)

df["Predicted_2025_Fatalities_RF"] = rf_regressor.predict(full_scaled).round(0).astype(int)

df["Predicted_Risk_Level"] = label_encoder.inverse_transform(
    rf_classifier.predict(full_scaled)
)

df["Predicted_2025_Fatalities_ANN"] = ann_model.predict(full_scaled).flatten().round(0).astype(int)

result_csv_path = os.path.join(BASE_DIR, "road_accident_prediction_results.csv")

df.to_csv(result_csv_path, index=False)


# ============================================================
# Final Output
# ============================================================

print("\nFiles saved successfully:")
print("H5 Model:", h5_path)
print("PKL Models:", pkl_path)
print("YAML Config:", yaml_path)
print("JSON Metrics:", json_path)
print("Prediction CSV:", result_csv_path)

print("\nSample Output:")
print(
    df[
        [
            state_col,
            target_col,
            "Risk_Level",
            "Predicted_2025_Fatalities_RF",
            "Predicted_2025_Fatalities_ANN",
            "Predicted_Risk_Level",
            "Recommended_Ambulances",
            "Recommended_Highway_Patrols",
            "Recommended_Trauma_Centers"
        ]
    ].head()
)
```
